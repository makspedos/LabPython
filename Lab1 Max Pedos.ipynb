```python
def foo(num):
    hasNumbers = False;
    for i in range(1,num+1):
        for j in range(i,num+1):
            if i*i+j*j == num:
                print("i=",i," j=",j)
                hasNumbers = True

    if hasNumbers == False:
        print("No numes")

try:
    a= int(input("Enter number"))
except(ValueError):
    print("Yut entered wrong value")
else:  
    foo(a)
    
```
