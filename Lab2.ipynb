```python
#Завдання 3 - матриці 3(A^2-B^2)-2AB
import time

start = time.time()


A = np.array([[4 , 2, 1],
              [3 ,-2 ,0],
              [0 , 1,2]])

B = np.array([[2,0,2],
              [5,-7,-2],
              [1,0,-1]])

result = 3*(A@A - B@B)- 2* A@B
stop = time.time()
print(f"Execution_Time:{(stop-start)}sec")
print(result)
```

```python
#Завдання 3 - матриці(2 спосіб) 3(A^2-B^2)-2AB
import time

start = time.time()


A = [[4,2,1] ,[3,-2,0] ,[0,1,2]]
B = [[2,0,2] ,[5,-7,-2] ,[1,0,-1]]
step1 = [[0 for i in range(len(A))] for i in range(len(A))]
step2 =[[0 for i in range(len(A))] for i in range(len(A))]
step3= [[0 for i in range(len(A))] for i in range(len(A))]
step4= [[0 for i in range(len(A))] for i in range(len(A))]
step5= [[0 for i in range(len(A))] for i in range(len(A))]
step6= [[0 for i in range(len(A))] for i in range(len(A))]
step7= [[0 for i in range(len(A))] for i in range(len(A))]


for i in range(len(A)):
    for j in range (len(A[i])):
        step1[i][j] = 0
        for z in range(len(A[j])):
             step1[i][j] += A[i][z] * A[z][j]    
 




for i in range(len(B)):
    for j in range (len(B[i])):
        step2[i][j] = 0
        for z in range(len(B[j])):
             step2[i][j] += B[i][z] * B[z][j]    
             
for i in range(len(B)):
    for j in range (len(B[i])):
        step3[i][j] += step1[i][j] - step2[i][j]
            
for i in range(len(B)):
    for j in range (len(B[i])):
        step4[i][j] += 3 *  step3[i][j]
            
for i in range(len(B)):
    for j in range (len(B[i])):
        step5[i][j] = 0
        for z in range(len(B[j])):
             step5[i][j] += A[i][z] * B[z][j]     

for i in range(len(B)):
    for j in range (len(B[i])):
        step6[i][j] += -2 *  step5[i][j]
                 
for i in range(len(B)):
    for j in range (len(B[i])):
        step7[i][j] += step4[i][j] + step6[i][j]
stop = time.time()
print(f"Execution_Time:{(stop-start)}sec")
print(step7)


```

```python
#Завдання 2 
a = np.array([[4 , 1 , 0 , -1], [1 , -3 , 4 , 0] , [0 , 3 , -2 , 4] , [1 , 0 , 2 , 4]])
b = np.array([-9 , 7 , 12 , 12 ])
"""First method"""
x = np.linalg.solve(a ,b )

"""Second method"""
matrix = np.linalg.inv(a)
x2 = np.matmul(matrix , b)

np.allclose(x , x2)
```
