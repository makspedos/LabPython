```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig= plt.figure()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
x = np.linspace(0, 20, 30)
axes[0].plot(31/34* x - 365/51, '--g' );
axes[0].set(xlim=(0,20), ylim=(-40, 80),
title='$â€“10.2 x_1 + 9.3 x_2 = 73$');


axes[1].plot(-158/53 * x + 1730/53 , 'b.' );
axes[1].set(xlim=(0,20), ylim=(-40, 80),
title='$5.3 x_1 + 15.8 x_2 = 173$');

axes[2].plot(-103/172 * x + 795/86, 'm-.');
axes[2].set(xlim=(0,20), ylim=(-40, 80),
title='$17.2 x_1 + 10.3 x_2 = 159$');

fig.tight_layout()
fig.savefig('my_figure.png', dpi=1000)
fig.savefig('my_figure.jpg', dpi=200)
fig.savefig('my_figure.svg', dpi=200)



```

```python
x1 = np.linspace(0, 20, 30)
plt.figure(figsize=(9, 9))

plt.plot(x1, 31/34* x - 365/51, '--g' , label='$f(1) = â€“10.2 x_1 + 9.3 x_2 = 73$')
plt.text(1 , -5,'$f(1) = â€“10.2 x_1 + 9.3 x_2 = 73$', fontsize = 12,rotation = 10);

plt.plot(x1 , -158/53 * x + 1730/53 , 'b.' , label='$f(2) = 5.3 x_1 + 15.8 x_2 = 173$' )
plt.text(3.5 , 10,'$f(2) = 5.3 x_1 + 15.8 x_2 = 173$', fontsize = 12,rotation = -25);

plt.plot(x1 , -103/172 * x + 795/86 , 'm-.' , label='f(3) =$17.2 x_1 + 10.3 x_2 = 159$')
plt.text(2 , 5,'$f(3) =$17.2 x_1 + 10.3 x_2 = 159$', fontsize = 12,rotation = -15);

plt.xlim(0,20)
plt.ylim(-40, 75);

plt.legend()
plt.title('All lines in one figure');
plt.xlabel("$x_1$")
plt.ylabel("$x_2$");

plt.scatter(10.22333096, 2.1644096, s=40 , color = 'red')
plt.arrow(x=10.22333096, y=2.1644096, dx=1, dy=30, width=.05) 
plt.annotate('The intersection of f(1) and f(2)', xy = (10, 35))

plt.scatter(10.85729379, 2.74243453, s=40 , color = 'red')
plt.arrow(x=10.85729379, y=2.74243453, dx=7, dy=20, width=.05) 
plt.annotate('The intersection of f(1) and f(3)', xy = (13, 24))

plt.scatter(9.82133812, 3.36280333, s=40 , color = 'red')
plt.arrow(x=9.82133812, y=3.36280333, dx=-5, dy=-20, width=.05) 
plt.annotate('The intersection of f(2) and f(3)', xy = (3, -19))


plt.show()

fig.savefig('my_figure.png', dpi=1000)
fig.savefig('my_figure.jpg', dpi=200)
fig.savefig('my_figure.svg', dpi=200)
```

```python
from numpy import linalg as LA
a = np.array([[-10.2 , 9.3] , [5.3 , 15.8]])
b = np.array([73 , 173])
print(LA.solve(a,b) , "\n")

a1 = np.array([[-10.2 , 9.3] , [17.2 , 10.3]])
b2 = np.array([73 , 159])
print(LA.solve(a1,b2) , "\n")
```
